```python
# import dependencies
import pandas as pd
import json
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
from webdriver_manager.chrome import ChromeDriverManager
```

```python
to_minor = {
    'A': 'F♯/G♭',
    'A♯/B♭': 'G',
    'B': 'G♯/A♭',
    'C': 'A',
    'C♯/D♭': 'A♯/B♭',
    'D': 'B',
    'D♯/E♭': 'C',
    'E': 'C♯/D♭',
    'F': 'D',
    'F♯/G♭': 'D♯/E♭',
    'G': 'E',
    'G♯/A♭': 'G♯/A♭',
}
```

```python
def get_songs(artist):
    
    songs = {
            'artist': [],
            'title': [],
            'bpm': [],
            'key': []
        }
    
    for artist in artists:
    
        # Try to scrape the site using the URL
        url = f'https://songbpm.com/@{artist}'

        # Retrieve page with the requests module
        html = requests.get(url)

        # Create BeautifulSoup object; parse with 'html.parser'
        soup = bs(html.text, 'html.parser')

        # Get article title and summary paragraph
        spans_keys = soup.find_all('span', class_='text-2xl text-gray-700 font-base sm:text-3xl')

        spans_songs = soup.find_all('span', class_='text-2xl font-bold text-gray-700 font-base sm:font-normal sm:text-3xl')

        paras_titles = soup.find_all('p', class_='pr-2 text-lg font-base sm:font-light sm:text-2xl')

        key = []

        for i, span in enumerate(spans_keys):
            if i % 2 == 0 and span.text != '':
                key.append((i,span.text))
                songs['key'].append(span.text)

        for k in key:
            songs['artist'].append(artist)
            songs['title'].append(paras_titles[int(k[0]/2)].text)
            songs['bpm'].append(spans_songs[int(k[0]/2)].text)

    songs_df = pd.DataFrame(songs)

    songs_df['relative_minor'] = songs_df['key'].map(to_minor)

    return songs_df

```

```python
artists = ['sahbabii', 'red-velvet']
songs_df = get_songs(artists)
songs_df
```
